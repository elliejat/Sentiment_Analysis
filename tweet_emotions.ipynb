# sentiment analysis

```python
#Data 
import pandas as pd
import numpy as np

#plots
import seaborn as sns
import matplotlib.pyplot as plt
import re
import nltk #for analyzing text

#Linear Regression Model Building Library
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('tweet_emotions.csv')
```

```python
df
```

```python
#change emotions to numbers
n = len(pd.unique(df['sentiment']))
print(n)
```

```python
from sklearn.preprocessing import OrdinalEncoder

ord_enc = OrdinalEncoder()
df["emotions"] = ord_enc.fit_transform(df[["sentiment"]])
df[["sentiment", "emotions"]].head(40000)
```

```python
#nlp on content column
df
```

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

corpus=[] #create an empty bag

#fill bag w/ words, goes thro each row (comment), breaks it down & add to baf
for i in range(40000):
    review = re.sub('[^a-zA-Z]',' ', df['content'][i])
    review= review.lower()
    review = review.split()
    ps=PorterStemmer()
    review=[ps.stem(word) for word in review if not word in set(stopwords.words('english'))] #take stem of every word in row
    review= ' '.join(review) #putting stemmed words together in each row
    corpus.append(review) #putting it back in dataset
    
```

```python
corpus
```

# Logistic Regression

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features= 1500)
X= cv.fit_transform(corpus).toarray()
y= df.iloc[:,3].values 
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y, test_size=0.25)
```

```python
#modeling
import sklearn.linear_model
model=sklearn.linear_model.LogisticRegression()
model.fit(X_train,y_train) #fit the logistical regression model on train data
y_pred=model.predict(X_test) #predict on x test data
```

```python
#accuracy
from sklearn.metrics import classification_report
matrix= classification_report(y_test,y_pred)
print(matrix)
```

# Knearest neighbor

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.25)
```

```python
from sklearn.neighbors import KNeighborsClassifier 
classifier = KNeighborsClassifier(n_neighbors = 5, metric ='minkowski', p=2)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)

#Accuracy
from sklearn.metrics import classification_report
matrix = classification_report(y_test,y_pred)
print(matrix)
```

# confusion matrix

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```
